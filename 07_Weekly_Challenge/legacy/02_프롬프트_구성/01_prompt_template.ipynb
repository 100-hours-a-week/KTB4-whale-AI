```python
from langchain_core.prompts import PromptTemplate

# === from_template으로 생성 (권장) ===
# {topic}과 {length}가 변수 자리표시자입니다.
# from_template은 문자열에서 {변수}를 자동으로 감지합니다.
prompt = PromptTemplate.from_template(
    "당신은 {role} 전문가입니다. {topic}에 대해 {length}자 이내로 설명하세요."
)

# === 변수 목록 확인 ===
# 어떤 변수가 필요한지 명시적으로 확인할 수 있습니다.
print(f"필요한 변수: {prompt.input_variables}")
print(f"템플릿 문자열: {prompt.template}")
```

```python
# === format(): 완성된 문자열을 반환 ===
result_str = prompt.format(role="교육", topic="어댑터즈", length="100")
print(f"format() 결과: {result_str}")
print(f"반환 타입: {type(result_str).__name__}")

print()

# === invoke(): StringPromptValue 객체를 반환 ===
result_obj = prompt.invoke({"role": "교육", "topic": "5단 분석법", "length": "200"})
print(f"invoke() 결과: {result_obj.text}") # type: ignore
print(f"반환 타입: {type(result_obj).__name__}")

print()
# === partial: role을 미리 고정 ===
# role 변수에 "교육"을 고정한 새 템플릿이 생성됩니다.
partial_prompt = prompt.partial(role="교육")
result = partial_prompt.format(topic="5단 분석법", length="200")
print(f"invoke() 결과: {result}")
print(f"partial 후 변수: {partial_prompt.input_variables}")
```

```python
# === 같은 템플릿, 다른 입력 ===
# 프롬프트 구조는 그대로 두고 변수만 바꿔서 반복 사용합니다.
topics = [
    {"role": "교육", "topic": "어댑터즈", "length": "100"},
    {"role": "교육", "topic": "5단 분석법", "length": "150"},
    {"role": "교육", "topic": "스타트업코드", "length": "80"},
]

for t in topics:
    result = prompt.format(**t)
    print(result)
```

```python
# === 직접 생성: input_variables를 명시적으로 지정 ===
prompt_direct = PromptTemplate(
    template="다음 문서를 참고하여 질문에 답하세요.\n\n[문서]\n{document}\n\n[질문]\n{question}",
    input_variables=["document", "question"]
)

print(f"필요한 변수: {prompt_direct.input_variables}")

result = prompt_direct.format(
    document="어댑터즈는 스타트업코드에서 제공하는 개발 책 서빙 서비스입니다.",
    question="어댑터즈는 무엇인가요?"
)
print(result)
```
