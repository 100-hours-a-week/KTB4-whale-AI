```python
from langchain_community.document_loaders import PyPDFLoader

# === 객체 생성 ===
loader = PyPDFLoader("파일경로.pdf")

# === 모든 페이지를 한 번에 ===
docs = loader.load()
# docs == [Document(페이지 0), Document(페이지 1), ...]
# 각 Document.metadata == {"source": "파일경로.pdf", "page": 0, ...}

# === 메모리 효율을 위해 페이지별로 ===
for doc in loader.lazy_load():
    print(doc.metadata["page"], doc.page_content[:50])
```

```python
# === Colab에 PDF 파일 업로드 ===
from google.colab import files
uploaded = files.upload()        # 파일 선택 다이얼로그가 열림 → sample_pdf.pdf 선택

pdf_path = next(iter(uploaded))  # 업로드한 파일명을 그대로 사용
print(f"업로드된 파일: {pdf_path}")
```

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader(pdf_path)

# === 한 번에 모든 페이지 읽기 ===
docs = loader.load()

print(f"전체 페이지 수: {len(docs)}")
print(f"\n첫 페이지 metadata: {docs[0].metadata}")
print(f"첫 페이지 본문(앞 80자): {docs[0].page_content[:80]}")
print(f"\n마지막 페이지 metadata: {docs[-1].metadata}")
print(f"마지막 페이지 본문(앞 80자): {docs[-1].page_content[:80]}")
```

```python
# 페이지를 하나씩 처리
loader = PyPDFLoader(pdf_path)

# === 페이지 단위로 하나씩 처리 ===
for doc in loader.lazy_load():
    page_num = doc.metadata["page"] + 1   # 사용자 표시는 1부터
    preview = doc.page_content[:40].replace("\n", " ")
    print(f"[페이지 {page_num}] {preview}...")
```

```python
# 특정 페이지만 골라 사용
loader = PyPDFLoader(pdf_path)
docs = loader.load()

# === metadata["page"] 기준으로 필터링 ===
# AI 커리큘럼 상세가 담긴 두 번째 페이지(page=1)만 사용
selected = [d for d in docs if d.metadata["page"] == 1]

for d in selected:
    print(f"[page {d.metadata['page']}] {d.page_content[:60]}...")
```
