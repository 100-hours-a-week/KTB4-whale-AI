```python
from langchain_core.documents import Document

# === 어댑터즈 소개 문서를 Document 객체로 만들기 ===
doc = Document(
    page_content="Adapterz(어댑터즈)는 스타트업코드에서 운영하는 개발 교재 서빙 서비스입니다. "
                 "5단 분석법이라는 독자적인 교수법으로 모든 교재를 작성합니다.",
    metadata={
        "source": "adapterz_intro.txt",
        "category": "service_intro",
        "author": "startupcode"
    }
)

print(f"page_content: {doc.page_content}")
print(f"metadata: {doc.metadata}")
print(f"출처: {doc.metadata['source']}")
print(f"타입: {type(doc).__name__}")
```

```python
# === 어댑터즈 관련 문서 조각 여러 개 ===
docs = [
    Document(
        page_content="어댑터즈는 스타트업코드에서 운영하는 개발 교재 서빙 서비스입니다.",
        metadata={"source": "adapterz_intro.txt", "section": "정의"}
    ),
    Document(
        page_content="어댑터즈의 모든 교재는 5단 분석법이라는 독자적인 교수법으로 작성됩니다.",
        metadata={"source": "adapterz_intro.txt", "section": "교수법"}
    ),
    Document(
        page_content="어댑터즈는 인공지능, 데이터 분석, 웹 개발, 인프라 분야의 교재를 제공합니다.",
        metadata={"source": "adapterz_intro.txt", "section": "분야"}
    ),
]

for i, d in enumerate(docs, 1):
    print(f"[{i}] section={d.metadata['section']} | {d.page_content}")
```

```python
# === 특정 section만 골라내기 ===
교수법_docs = [d for d in docs if d.metadata["section"] == "교수법"]

for d in 교수법_docs:
    print(d.page_content)
```
