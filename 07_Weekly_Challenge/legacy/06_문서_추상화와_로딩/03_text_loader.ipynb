```python
# === 어댑터즈 소개 텍스트를 파일로 저장 (실습용) ===
content = (
    "Adapterz(어댑터즈)는 스타트업코드에서 운영하는 개발 교재 서빙 서비스입니다.\n"
    "어댑터즈의 모든 교재는 5단 분석법이라는 독자적인 교수법으로 작성됩니다.\n"
    "5단 분석법은 기술 용어를 일반 명사와 고유 명사로 나누어 설명하고, "
    "사용 이유와 방법, 다른 기술과의 비교까지 체계적으로 정리하는 방식입니다.\n"
    "어댑터즈는 인공지능, 데이터 분석, 웹 개발, 인프라 분야의 교재를 제공합니다."
)

with open("adapterz_intro.txt", "w", encoding="utf-8") as f:
    f.write(content)
```

```python
from langchain_community.document_loaders import TextLoader

# === TextLoader 객체 생성 ===
loader = TextLoader("adapterz_intro.txt", encoding="utf-8")

# === 파일을 한 번에 읽기 ===
docs = loader.load()

print(f"읽어 온 Document 수: {len(docs)}")
print(f"타입: {type(docs[0]).__name__}")
print(f"\n=== page_content ===")
print(docs[0].page_content)
print(f"\n=== metadata ===")
print(docs[0].metadata)
```

```python
loader = TextLoader("adapterz_intro.txt", encoding="utf-8")

# === lazy_load는 제너레이터를 반환 ===
for i, doc in enumerate(loader.lazy_load(), 1):
    print(f"[{i}] source={doc.metadata['source']}")
    print(f"    {doc.page_content[:40]}...")
```
