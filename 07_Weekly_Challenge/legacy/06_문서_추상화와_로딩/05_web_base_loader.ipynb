```python
from langchain_community.document_loaders import WebBaseLoader

# === 어댑터즈 메인 페이지 ===
url = "https://academy.adapterz.kr"

loader = WebBaseLoader(url)
docs = loader.load()

print(f"Document 수: {len(docs)}")
print(f"본문 길이: {len(docs[0].page_content)}자")
print(f"\n=== metadata ===")
print(docs[0].metadata)
print(f"\n=== page_content (앞 150자) ===")
print(docs[0].page_content[:150])
```

```python
# === 여러 페이지를 한 번에 ===
urls = [
    "https://academy.adapterz.kr",
    "https://academy.adapterz.kr/fullstack",
    "https://academy.adapterz.kr/faq",
]

loader = WebBaseLoader(urls)
docs = loader.load()

print(f"전체 Document 수: {len(docs)}")
for d in docs:
    print(f"- [{len(d.page_content)}자] {d.metadata['title']} ({d.metadata['source']})")
```

```python
import bs4

# === <article> 태그 영역만 파싱 ===
# 사이트마다 본문 영역의 태그/클래스가 다르므로, 실제 사용 시에는
# 브라우저 개발자 도구로 본문 컨테이너의 태그·클래스를 먼저 확인합니다.
loader = WebBaseLoader(
    web_paths=["https://academy.adapterz.kr"],
    bs_kwargs={
        "parse_only": bs4.SoupStrainer("article")
    }
)
docs = loader.load()

print(f"본문 영역만 추출된 텍스트 길이: {len(docs[0].page_content)}자")
print(f"앞 100자: {docs[0].page_content[:100]}")
```
