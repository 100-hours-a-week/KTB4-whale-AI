# 세부 과제 4 - GridSearch, RandomSearch로 하이퍼파라미터 튜닝하기

## 내용 요약

- 본 과제는 가상으로 만든 무작위 이미지 데이터를 사용해 두 가지 하이퍼파라미터를 튜닝하기
- 두 가지 대표적인 하이퍼파라미터 탐색 방법인 GridSearch, RandomSearch를 각각 구현하여 실행
- 그 결과를 비교 분석하여, 두 방법의 특징, 장단점, 실제 동작의 차이 등을 이해하는 것을 목적으로 삼음

## 주요 개념 정리

### 하이퍼파라미터
- 모델이 학습하기 전에 사용자가 설정해주어야 하는 값
- 모델이 무엇을 배울지는 데이터가 결정하지만, 어떻게 배울지는 하이퍼파라미터가 결정

|구분|모델 파라미터 (Model Parameters)|하이퍼파라미터 (Hyperparameters)|
|-|-|-|
|학습 여부|학습 과정에서 자동으로 업데이트됨|사용자가 직접 설정해야 함|
|예시|가중치(W), 편향(b)|learning_rate, batch_size, epochs, dropout 비율, layer 수 등|
|역할|데이터로부터 패턴을 학습|학습 방식과 모델 구조를 제어|

### 튜닝의 목적
- 같은 모델 구조라도 하이퍼파라미터 값에 따라 성능이 크게 달라질 수 있기 때문
    - learning rate가 너무 크면, 학습이 발산하거나 불안정
    - learning rate가 너무 작으면, 학습이 느리거나 지역 최적해에 갇힘
    - batch size가 너무 크면, 일반화 성능이 떨어질 수 있음
    - batch size가 너무 작으면, 학습이 불안정하고 시간이 오래 걸림

- 좋은 하이퍼파라미터를 찾는 것은 모델 성능을 결정하는 가장 중요한 요소 중 하나
- 따라서 실무에서는 하이퍼파라미터 튜닝을 체계적으로 수행해야 함

### 하이퍼파라미터 튜닝의 대표적인 두 가지 방법
- GridSearch, RandomSearch가 가장 기본적이고 대표적인 두 가지 방법

|방법|정의|특징|장점|단점|
|-|-|-|-|-|
|GridSearch|미리 정해놓은 값들의 모든 조합을 하나씩 전부 시도하는 방법|체계적, 결정론적|모든 조합을 빠짐없이 확인 가능|조합 수가 많아지면 계산 비용이 폭발적으로 증가|
|RandomSearch|하이퍼파라미터 공간에서 무작위로 일정 횟수만큼 샘플링하여 시도|확률적, 비결정론적|계산 비용이 상대적으로 적음|운에 따라 결과가 달라질 수 있음|

- 요약
    - GridSearch: “모든 경우의 수를 다 확인하자”는 접근
    - RandomSearch: “중요한 영역을 운 좋게 빨리 찾자”는 접근(Bergstra & Bengio, 2012 논문 참고)

### 비교 분석 내용 (실제 실험 결과 기반)

- 실험 결과 요약표
    |항목|GridSearch|RandomSearch|비고|
    |-|-|-|-|
    |시도한 Trial 수|9|5|-|
    |Best Test Accuracy|0.130|0.125|GridSearch 근소 우위|
    |Best Parameters|learning_rate=0.01, batch_size=64|learning_rate≈0.0607, batch_size=16|-|
    |성능 범위|0.090 ~ 0.130|0.120 ~ 0.125|GridSearch가 더 넓게 분포|
    |높은 LR에서의 불안정성|다수 발생 (초기 loss 11~351 폭발)|상대적으로 적음|RandomSearch 우위|
    |탐색 방식|모든 조합을 체계적으로 실행|무작위 샘플링|-|
    |계산 효율성|낮음|높음|RandomSearch 우위|

- 주요 분석 포인트
    - 성능: GridSearch가 0.13으로 약간 더 높았지만, 데이터가 완전 무작위였기 때문에 큰 의미없음
    - 효율성: RandomSearch가 5번의 시도만으로 GridSearch(9번)와 비슷한 성능을 달성
    - 안정성: GridSearch는 learning_rate=0.1 구간에서 학습이 크게 불안정해지는 현상이 자주 발생(RandomSearch는 이런 극단적인 bad region에 덜 노출)
    - 결론: 이번 실험에서는 RandomSearch가 계산 효율성과 안정성 측면에서 더 우수한 결과 확인

### 최종 회고록
- 이번 과제를 통해 하이퍼파라미터 튜닝에 대한 개념과 실제 구현 경험을 얻을 수 있었다.

- 가장 인상 깊었던 점은 같은 모델이라도 하이퍼파라미터에 따라 학습 안정성과 최종 성능이 크게 달라진다는 점이다. 특히 GridSearch를 실행하면서 높은 learning_rate에서 loss가 폭발적으로 증가하는 현상을 직접 관찰하며, 왜 무작정 높은 값을 사용하면 안 되는지 체감할 수 있었다.
- 또한 GridSearch와 RandomSearch를 직접 비교해보니, 단순히 “더 많은 조합을 시도하는 것”이 항상 좋은 것이 아니라는 점을 알게 되었다. RandomSearch는 적은 시도 횟수로도 준수한 결과를 낼 수 있었고, 이는 실무에서 제한된 시간과 자원 내에서 하이퍼파라미터를 탐색할 때 효율적이라는 중요한 시사점을 주었다.
- 다만 이번 실험은 데이터가 완전 무작위였기 때문에, 실제 이미지 데이터(CIFAR-10 등)에서는 두 방법의 차이가 더 극명하게 드러날 것이라고 예상된다. 다음에는 실제 데이터셋을 사용하거나, KerasTuner 같은 고급 도구를 활용하여 Bayesian Optimization까지 경험해보고 싶었다.

- 이번 과제를 통해 하이퍼파라미터 튜닝은 **모델의 성능과 학습 안정성**을 좌우하는 설계라는 것을 알 수 있었다. 프로젝트를 진행할 때 무작적 기본값을 사용하기보다, 체계적인 탐색 과정을 거치는 습관을 들여야겠다고 생각했다.
