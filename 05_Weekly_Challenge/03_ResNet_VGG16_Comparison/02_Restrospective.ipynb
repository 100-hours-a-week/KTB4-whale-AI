- 주요 관찰 및 분석
    1. 정확도 측면
        - 두 모델 모두 Pretrained를 사용한 덕분에 10 Epoch 만에 81% 이상의 성능을 달성했다.
        - ResNet18가 VGG16보다 파라미터 수가 훨씬 적음에도 불구하고, 정확도에서 큰 차이를 보이지 않았다.
        - 이는 Pretrained Weight의 영향이 크다는 것을 보여준다. (Scratch 학습이었다면 차이가 더 컸을 가능성이 높음)
    2. 속도 및 효율성 측면
        - ResNet18이 압도적으로 빠르다. 총 훈련 시간 기준 약 3.5배 차이.
        - VGG16은 구조적으로 Fully Connected Layer가 매우 크고, 연산량이 많기 때문에 같은 환경에서 불리하다.
        - 특히 Colab T4와 같은 중저사양 GPU에서는 ResNet 계열이 실용적이다.
    3. 수렴 속도
        - VGG16은 Epoch이 진행될수록 Loss가 더 빠르게 떨어지는 경향을 보였다.
        - ResNet18은 상대적으로 안정적이면서도 빠르게 수렴했다.
        - 두 모델 모두 Pretrained 덕분에 초기 Loss가 낮게 시작했다.
    4. 모델 선택 관점
        - 정확도만 본다면 두 모델의 차이가 크지 않다.
        - 훈련 속도, 메모리 효율성, 실험 반복성을 고려하면 **ResNet18이 압도적으로 유리**하다.
        - VGG16은 구조가 단순하고 해석이 쉽다는 장점이 있지만, 현대적인 실험 환경에서는 ResNet 계열이 더 선호된다.

- 결론 및 배운 점
    - Pretrained Weight를 사용하면 모델 구조에 따른 성능 차이가 줄어든다.
    - 같은 Pretrained 환경에서는 계산 효율성이 더 중요한 선택 기준이 될 수 있다.
    - VGG16은 파라미터가 많아서 overfitting 위험이 더 클 수 있지만, Pretrained를 사용하면 이 위험이 상당 부분 완화된다.
    - 실제 프로젝트에서는 정확도뿐만 아니라 **훈련 속도와 자원 효율성도 중요한 평가 지표**임을 확인할 수 있었다.
