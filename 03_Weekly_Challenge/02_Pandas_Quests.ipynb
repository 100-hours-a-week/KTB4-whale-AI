## 필터링

```python
# 1번 퀘스트
# 주어진 데이터프레임에서 나이가 30 이상인 행만 필터링하기
data = {'이름': ['홍길동', '김철수', '박영희', '이순신', '강감찬'],
        '나이': [25, 30, 35, 40, 45],
        '도시': ['서울', '부산', '서울', '대구', '부산']}

df = pd.DataFrame(data)
filtered = df[df['나이'] >= 30]
print(filtered)
```

```python
# 2번 퀘스트
# 주어진 데이터프레임에서 도시가 '서울'이거나 점수가 80점 이상인 데이터를 필터링하기
data = {'이름': ['홍길동', '김철수', '박영희', '이순신', '강감찬'],
        '나이': [25, 30, 35, 40, 45],
        '도시': ['서울', '부산', '서울', '대구', '부산'],
        '점수': [85, 90, 75, 95, 80]}

df = pd.DataFrame(data)
# filtered = df[(df['도시'] == '서울') | (df['점수'] >= 80)] # 논리 연산자
# filtered = df.query("도시 == '서울' or 점수 >= 80") # query
# filtered = df[(df['도시'].isin(['서울'])) | (df['점수'] >= 80)] # isin()
# filtered = df[(df['도시'].str.contains('서울')) | (df['점수'] >= 80)] # str.contains()
filtered = df[lambda x: (x['도시'] == '서울') | (x['점수'] >= 80)] # lambda
print(filtered)
```

```python
# 3번 퀘스트
# 주어진 프레임에서 query()를 사용해 나이가 35 이상이고 점수가 80점 초과인 데이터를 필터링하기
data = {'이름': ['홍길동', '김철수', '박영희', '이순신', '강감찬'],
        '나이': [25, 30, 35, 40, 45],
        '도시': ['서울', '부산', '서울', '대구', '부산'],
        '점수': [85, 90, 75, 95, 80]}

df = pd.DataFrame(data)
filtered = df.query("나이 >= 35 and 점수 > 80")
print(filtered)
```

## 그룹화

```python
# 1번 퀘스트
# 주어진 데이터프레임에서 groupby를 사용해 부서별로 급여의 합계를 구하기
data = {
    '이름': ['홍길동', '김철수', '박영희', '이순신'],
    '부서': ['영업', '영업', '인사', '인사'],
    '급여': [5000, 5500, 4800, 5100]
}

df = pd.DataFrame(data)
grouped = df.groupby('부서') # 분할
sums = grouped['급여'].sum() # 적용 및 결합
print(sums)
```

```python
# 2번 퀘스트
# 주어진 데이터프레임에서 groupby와 agg를 사용하여 부서별 급여의 합계와 평균을 계산하기
data = {
    '이름': ['홍길동', '김철수', '박영희', '이순신', '강감찬', '신사임당'],
    '부서': ['영업', '영업', '인사', '인사', 'IT', 'IT'],
    '급여': [5000, 5500, 4800, 5100, 6000, 6200]
}

df = pd.DataFrame(data)
grouped = df.groupby('부서')
calculated = grouped['급여'].agg(['sum', 'mean'])
print(calculated)
```

```python
# 3번 퀘스트
# 부서별로 데이터를 그룹화한 뒤, 각 부서의 평균 급여가 5000 이상인 경우만 필터링하기
data = {
    '이름': ['홍길동', '김철수', '박영희', '이순신', '강감찬', '신사임당'],
    '부서': ['영업', '영업', '인사', '인사', 'IT', 'IT'],
    '급여': [5000, 5500, 4800, 5100, 6000, 6200]
}

df = pd.DataFrame(data)
grouped = df.groupby('부서')
df['급여_평균'] = grouped['급여'].transform('mean')
filtered = df[df['급여_평균'] >= 5000]
print(filtered)
```

## 병합

```python
# 1번 퀘스트
# 두 개의 데이터프레임을 고객ID를 기준으로 내부 조인하여 공통된 데이터를 병합하기
df1 = pd.DataFrame({'고객ID': [1, 2, 3], '이름': ['홍길동', '김철수', '이영희']})
df2 = pd.DataFrame({'고객ID': [2, 3, 4], '구매액': [10000, 20000, 30000]})
# merged = pd.merge(df1, df2) # on은 default로 공통 column 사용 (현재 공통 컬럼 "고객 ID")
# merged = pd.merge(df1, df2, on='고객ID') # how는 default로 inner
# merged = pd.merge(df1, df2, on='고객ID', how='left') # left, right에 빈 값 있으면 NaN
# merged = pd.merge(df1, df2, on='고객ID', how='right') # right, left에 빈 값 있으면 NaN
# merged = pd.merge(df1, df2, on='고객ID', how='outer') # 합집합
merged = pd.merge(df1, df2, on='고객ID', how='inner') # 교집합
print(merged)
```

```python
# 2번 퀘스트
# 데이터프레임을 왼쪽 기준으로 병합을 수행하고, 구매액이 없을 경우 NaN을 유지하기
df1 = pd.DataFrame({'고객ID': [1, 2, 3], '이름': ['홍길동', '김철수', '이영희']})
df2 = pd.DataFrame({'고객ID': [2, 3, 4], '구매액': [15000, 25000, 35000]})
merged = pd.merge(df1, df2, how='left')
print(merged)
```

```python
# 3번 퀘스트
# 두 개 이상의 열을 기준으로 병합하고
# 중복 열의 이름을 구분하도록 접미사(suffixes)를 지정하기
df1 = pd.DataFrame({
    '고객ID': [1, 2, 3],
    '도시': ['서울', '부산', '대전'],
    '구매액': [10000, 20000, 30000]
})

df2 = pd.DataFrame({
    '고객ID': [1, 2, 3],
    '도시': ['서울', '부산', '광주'],
    '구매액': [15000, 25000, 35000]
})

# 고객ID & 도시
merged = pd.merge(df1, df2, on=['고객ID', '도시'], suffixes=('_기존', '_신규'))
print(merged)
# 고객ID & 구매액 (데이터 없음)
# merged = pd.merge(df1, df2, on=['고객ID', '구매액'], suffixes=('_기존', '_신규'))
# print(merged)
# 도시 & 구매액 (데이터 없음)
# merged = pd.merge(df1, df2, on=['도시', '구매액'], suffixes=('_기존', '_신규'))
# print(merged)
```

## 결측치 처리

```python
# 1번 퀘스트
# 주어진 데이터프레임에서 결측치를 탐지하고, 열별 결측치 개수를 출력하기
data = {'이름': ['홍길동', '김철수', np.nan, '이영희'],
        '나이': [25, np.nan, 30, 28],
        '성별': ['남', '남', '여', np.nan]}

df = pd.DataFrame(data)
print(df.isnull()) # (결측치 탐지) 한계 - 데이터가 많을 경우, 탐지하기 어려움
print(df.info()) # (결측치 탐지) RangeIndex: 4, 3 non-null이므로 결측치 존재
print(df.isnull().sum()) # 열별 결측치 개수
```

```python
# 2번 퀘스트
# 주어진 데이터프레임에서 결측치가 포함된 행을 삭제하고
# 새로운 데이터프레임을 출력하기
data = {'이름': ['홍길동', '김철수', np.nan, '이영희'],
        '나이': [25, np.nan, 30, 28],
        '성별': ['남', '남', '여', np.nan]}

df = pd.DataFrame(data)
droped = df.dropna() # default가 axis=0이고, NaN이 있는 행을 제거함을 의미
droped_col = df.dropna(axis=1) # NaN이 있는 열을 제거함을 의미
droped_subset = df.dropna(subset=['이름'])
# droped_subset = df.dropna(subset=['이름'], axis=1) # 동작 불가(subset은 행 삭제용이므로, axis=1 사용 불가)
print(droped)
print(droped_col)
print(droped_subset)
```

```python
# 3번 퀘스트
# 결측치가 포함된 '나이' 열을 평균값으로 대체하고, 새로운 데이터프레임을 출력하기
data = {'이름': ['홍길동', '김철수', np.nan, '이영희'],
        '나이': [25, np.nan, 30, 28],
        '성별': ['남', '남', '여', np.nan]}

df = pd.DataFrame(data)

# 방법 1. 전체 데이터프레임에서 접근
mean = df['나이'].mean()
filled_mean = df.fillna(value={'나이': mean})
print("✅ 전체 데이터프레임에서 접근하여 '나이'열을 평균값으로 대체하고 새로운 데이터프레임을 출력하기")
print(filled_mean)

# 방법 2. 특정 열에서 접근
df_copy = df.copy()
df_copy['나이'] = df['나이'].fillna(mean)
print("\n✅ 특정 열 접근하여 '나이'열을 평균값으로 대체하고 새로운 데이터프레임을 출력하기")
print(df_copy)
```

## 피벗

```python
# 1번 퀘스트
# 주어진 데이터프레임을 피벗 함수로 날짜를 행 인덱스로, 제품을 열로, 판매량을 값으로 설정하기
data = {
    '날짜': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'],
    '제품': ['A', 'B', 'A', 'B'],
    '판매량': [100, 200, 150, 250]
}

df = pd.DataFrame(data)
pivoted_df = df.pivot(index='날짜', columns='제품', values='판매량') # 단일 피벗
print(pivoted_df)
```

```python
# 2번 퀘스트
# pivot_table을 사용해 주어진 데이터프레임에서
# 카테고리를 행 인덱스로, 제품을 열로, 판매량의 합계를 출력하기
data = {
    '카테고리': ['전자', '가전', '전자', '가전'],
    '제품': ['A', 'B', 'A', 'B'],
    '판매량': [100, 200, 150, 250]
}

df = pd.DataFrame(data)
df_pivot_table = df.pivot_table(index='카테고리', columns='제품', values='판매량', aggfunc='sum')
print(df_pivot_table)
```

```python
# 3번 퀘스트
# 주어진 데이터프레임에서 여러 값을 동시에 피벗하여
# 날짜를 행으로, 제품을 열로, 판매량과 이익을 동시에 피벗하기
data = {
    '날짜': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'],
    '제품': ['A', 'B', 'A', 'B'],
    '판매량': [100, 200, 150, 250],
    '이익': [20, 50, 30, 60]
}

df = pd.DataFrame(data)
pivoted_df = df.pivot(index='날짜', columns='제품', values=['판매량', '이익'])
print(pivoted_df)
```

## 중복 제거

```python
# 1번 퀘스트
# 주어진 데이터프레임에서 기본적으로 중복된 행을 제거하기
data = {
    '이름': ['김철수', '이영희', '김철수', '박민수'],
    '나이': [25, 30, 25, 40],
    '성별': ['남', '여', '남', '남']
}

df = pd.DataFrame(data)
droped_duplcate_df = df.drop_duplicates()
print(droped_duplcate_df)
```

```python
# 2번 퀘스트
# 주어진 데이터프레임에서 특정 열을 기준으로 중복을 제거한 후, 중복 여부를 확인하기
data = {
    '제품': ['노트북', '태블릿', '노트북', '스마트폰'],
    '가격': [1500000, 800000, 1500000, 1000000],
    '카테고리': ['전자기기', '전자기기', '전자기기', '전자기기']
}

df = pd.DataFrame(data)
# 0. (전) 중복 여부 확인
duplicates = df.duplicated(subset='카테고리')
print("✅ 0. (전) 중복 여부 확인")
print(duplicates)

# 1. 중복 제거
# droped_duplcate_df = df.drop_duplicates(subset='제품')
# droped_duplcate_df = df.drop_duplicates(subset='가격')
droped_duplcate_df = df.drop_duplicates(subset='카테고리')
print("\n✅ 1. 중복 제거")
print(droped_duplcate_df)

# 2. (후) 중복 여부 확인
duplicates = droped_duplcate_df.duplicated(subset='카테고리')
print("\n✅ 2. (후) 중복 여부 확인")
print(duplicates)
```

```python
# 3번 퀘스트
# 주어진 데이터프레임에서 중복된 모든 행을 삭제한 후
# 남은 데이터를 저장하고 다시 불러오기
data = {
    '학생': ['김민수', '박지현', '김민수', '이정훈'],
    '성적': [90, 85, 90, 88],
    '학교': ['A고', 'B고', 'A고', 'C고']
}

df = pd.DataFrame(data)
droped_duplicates_df = df.drop_duplicates(keep=False) # 중복된 모든 행 삭제

df.to_csv('students.csv', index=False) # 남은 데이터 저장

df_loaded = pd.read_csv('students.csv') # 다시 불러오기
print(df_loaded)
```

### 고유 명사
- 데이터를 특정 기준에 따라 그룹화하여 집계, 변환, 필터링 등의 연산을 수행하는 기능

- 그룹화는 "분할-적용-결합(Split-Apply-Combine)" 패턴
    - 분할(Split): 데이터를 특정 기준(열 또는 인덱스 값)에 따라 여러 개의 그룹으로 나눔
    - 적용(Apply): 그룹별로 다양한 연산(집계, 변환, 필터링)을 적용
    - 결합(Combine): 적용된 결과를 다시 하나의 데이터프레임 또는 시리즈로 결합

- 그룹화의 주요 개념
    - 그룹 키(Group Key): 데이터를 그룹화하는 기준이 되는 열(column) 또는 인덱스 값. 단일 키뿐만 아니라 복수의 키를 조합하여 그룹화 가능
    - 집계(Aggregation): 그룹 내에서 특정 연산(예: 합계, 평균, 최댓값)을 수행하여 요약된 결과를 제공
    - 변환(Transformation): 각 그룹별로 변환 연산을 수행하여 원래 데이터의 크기를 유지하며 값을 변경

```python
#################################################################################
# 추천 데이터셋: 회사 직원 정보
data = {
    '이름': ['홍길동', '김철수', '박영희', '이순신', '강감찬', '유관순', '신사임당', '세종대왕'],
    '부서': ['영업', '개발', '영업', '개발', '마케팅', '영업', '개발', '마케팅'],
    '직급': ['대리', '과장', '대리', '부장', '대리', '과장', '대리', '부장'],
    '급여': [4200, 5800, 4500, 7200, 3800, 5500, 4600, 6800],
    '근속연수': [3, 7, 4, 12, 2, 6, 5, 15],
    '입사연도': [2022, 2019, 2021, 2015, 2023, 2020, 2021, 2018]
}
```

```python
# DataFrame 생성 및 groupby 기능 활용
# =================== DataFrame 생성 ===================
df = pd.DataFrame(data)
print("✅ 회사 직원 정보 DataFrame 생성")
print(df)

# =================== groupby 기능 활용 ===================
# 1. 단일 열 기준 그룹화
grouped_1 = df.groupby('부서') # 분할: 부서 기준으로 그룹 분할
meaned = grouped_1['급여'].mean().round(1) # 적용 및 결합: 급여(숫자) 컬럼에 평균 함수 적용(결합은 자동)
renamed_1 = meaned.reset_index().rename(columns={'급여': "평균_급여"})
print("\n✅ [1. 단일 열 기준 그룹화] groupby로 부서별 급여 평균 산출")
print(renamed_1)

# 2. 여러 열 기준 그룹화
grouped_2 = df.groupby(['부서', '이름']) # 분할: 부서, 이름 기준으로 그룹 분할
sums = grouped_2['급여'].sum()# 적용 및 결합: 급여(숫자) 컬럼에 총계 함수 적용
renamed_2 = sums.reset_index().rename(columns={'급여': '급여_총계'})
print("\n✅ [2. 여러 열 기준 그룹화] groupby로 부서 및 이름별 급여 총계 산출")
print(renamed_2)

# 3. 그룹별 여러 집계 연산
grouped_1 = df.groupby('부서') # 분할: 부서 기준으로 그룹 분할
aggregated = grouped_1['급여'].agg(['sum', 'mean', 'max', 'min', 'size', 'count', 'median']) # 적용 및 결합
renamed_3 = aggregated.reset_index().rename(columns={
    'sum': '총_급여',
    'mean': '평균_급여',
    'max': '최고_급여',
    'min': '최저_급여',
    'size': '인원_수',
    'count': '유효_데이터_수',
    'median': '중앙값_급여'
})
print("\n✅ [3. 그룹 기준 여러 집계 연산] groupby로 부서별 급여 집계 연산 산출")
print(renamed_3)
```

## (필수, 추가 챌린지) DataFrame 생성 및 필터링 기능 사용

```python
# DataFrame 생성 및 filtering 기능 활용
# =================== DataFrame 생성 ===================
df = pd.DataFrame(data)
print("✅ 회사 직원 정보 DataFrame 생성")
print(df)

# =================== filtering 기능 활용 ===================
# 1. 그룹화 후 필터링
grouped_1 = df.groupby('부서') # 분할: 부서 기준으로 그룹 분할
filtered = grouped_1.filter(lambda x: x['급여'].mean() > 5000) # 적용 및 결합 # type: ignore
print("\n✅ [1. 그룹화 후 필터링] groupby로 부서별 급여 평균 중 5000 초과 직원 산출")
print(filtered)

# 2. 그룹별 변환: 그룹별 계산 결과를 원래 행에 기존 데이터프레임에 매칭해 반환하는 기능
grouped_1 = df.groupby('부서')
df['급여_평균'] = grouped_1['급여'].transform('mean').round(1)
filtered = df[df['급여_평균'] > 5000]
print("\n✅ [2. 그룹별 변환 후 필터링] groupby로 부서별 급여 평균 컬럼 변환하여 기존 데이터프레임에 매칭")
print(filtered)
```
