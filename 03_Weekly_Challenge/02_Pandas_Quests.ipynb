# Pandas 미니퀘스트

```python
import pandas as pd
```

## 시리즈(Series)

```python
# 1번 퀘스트
#  리스트를 사용하여 시리즈를 생성하고, 시리즈의 인덱스를 확인하기
array = [5, 10, 15, 20]
series = pd.Series(array)
print(series)
print(f"series.index: {series.index}")
```

```python
# 2번 퀘스트
# 다음 딕셔너리를 이용해 Series를 생성하고,
# 인덱스를 활용하여 b의 값을 출력하기
data = {'a': 100, 'b': 200, 'c': 300}
series = pd.Series(data)
print(series['b'])
```

```python
# 3번 퀘스트
# 다음 시리즈에 대해 결측값을 확인하고
# 모든 결측값을 0으로 채운 후
# 시리즈의 값을 출력하기
series = pd.Series([1, 2, None, 4, None, 6])
# series[series == None] = 0 (왜?)
filled_series = series.fillna(0)
print(series.values)
print(series)
print(filled_series)
```

## 데이터프레임(DataFrame)

```python
# 1번 퀘스트
# 다음 데이터프레임에서 열의 이름을 출력하는 코드를 작성하라.
data = {'이름': ['홍길동', '김철수', '박영희'], 
        '나이': [25, 30, 28], 
        '성별': ['남', '남', '여']}
df = pd.DataFrame(data)
print(df.columns.to_list())
print(list(df.columns))
```

```python
# 2번 퀘스트
# 다음 데이터프레임에서 나이 열을 기준으로
# 오름차순으로 정렬된 새로운 데이터프레임을 생성하고 출력하기
data = {'이름': ['홍길동', '김철수', '박영희'], 
        '나이': [25, 30, 28], 
        '성별': ['남', '남', '여']}
df = pd.DataFrame(data)
sorted_df = df.sort_values(by='나이', ascending=True)
print(sorted_df)
```

```python
# 3번 퀘스트
# 각 학생의 총점을 계산하는 새로운 열을 추가한 후
# 총점이 250점 이상인 학생들만 포함된 데이터프레임을 생성하기
data = {'이름': ['홍길동', '김철수', '박영희', '이순신'], 
        '국어': [85, 90, 88, 92], 
        '영어': [78, 85, 89, 87], 
        '수학': [92, 88, 84, 90]}
df = pd.DataFrame(data)

# 방법 1. 하드코딩
# df['총점'] = df['국어'] + df['영어'] + df['수학']

# 방법 2. 이름 열을 제외한 컬럼 선택
# score_columns = df.columns[df.columns != '이름']
# df['총점'] = df[score_columns].sum(axis=1)

# 방법 3. 숫자형 열만 있는 컬럼 선택
df['총점'] = df.select_dtypes(include='number').sum(axis=1)

high_total = df[df['총점'] >= 250]
print(high_total)
```
