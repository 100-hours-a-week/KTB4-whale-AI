```python
#################################################################################
# 추천 데이터셋: 회사 직원 정보
data = {
    '이름': ['홍길동', '김철수', '박영희', '이순신', '강감찬', '유관순', '신사임당', '세종대왕'],
    '부서': ['영업', '개발', '영업', '개발', '마케팅', '영업', '개발', '마케팅'],
    '직급': ['대리', '과장', '대리', '부장', '대리', '과장', '대리', '부장'],
    '급여': [4200, 5800, 4500, 7200, 3800, 5500, 4600, 6800],
    '근속연수': [3, 7, 4, 12, 2, 6, 5, 15],
    '입사연도': [2022, 2019, 2021, 2015, 2023, 2020, 2021, 2018]
}
```

## (필수, 추가 챌린지) DataFrame 생성 및 groupby 기능 활용

### 고유 명사
- 데이터를 특정 기준에 따라 그룹화하여 집계, 변환, 필터링 등의 연산을 수행하는 기능

- 그룹화는 "분할-적용-결합(Split-Apply-Combine)" 패턴
    - 분할(Split): 데이터를 특정 기준(열 또는 인덱스 값)에 따라 여러 개의 그룹으로 나눔
    - 적용(Apply): 그룹별로 다양한 연산(집계, 변환, 필터링)을 적용
    - 결합(Combine): 적용된 결과를 다시 하나의 데이터프레임 또는 시리즈로 결합

- 그룹화의 주요 개념
    - 그룹 키(Group Key): 데이터를 그룹화하는 기준이 되는 열(column) 또는 인덱스 값. 단일 키뿐만 아니라 복수의 키를 조합하여 그룹화 가능
    - 집계(Aggregation): 그룹 내에서 특정 연산(예: 합계, 평균, 최댓값)을 수행하여 요약된 결과를 제공
    - 변환(Transformation): 각 그룹별로 변환 연산을 수행하여 원래 데이터의 크기를 유지하며 값을 변경

```python
# DataFrame 생성 및 groupby 기능 활용
# =================== DataFrame 생성 ===================
df = pd.DataFrame(data)
print("✅ 회사 직원 정보 DataFrame 생성")
print(df)

# =================== groupby 기능 활용 ===================
# 1. 단일 열 기준 그룹화
grouped_1 = df.groupby('부서') # 분할: 부서 기준으로 그룹 분할
meaned = grouped_1['급여'].mean().round(1) # 적용 및 결합: 급여(숫자) 컬럼에 평균 함수 적용(결합은 자동)
renamed_1 = meaned.reset_index().rename(columns={'급여': "평균_급여"})
print("\n✅ [1. 단일 열 기준 그룹화] groupby로 부서별 급여 평균 산출")
print(renamed_1)

# 2. 여러 열 기준 그룹화
grouped_2 = df.groupby(['부서', '이름']) # 분할: 부서, 이름 기준으로 그룹 분할
sumed = grouped_2['급여'].sum()# 적용 및 결합: 급여(숫자) 컬럼에 총계 함수 적용
renamed_2 = sumed.reset_index().rename(columns={'급여': '급여_총계'})
print("\n✅ [2. 여러 열 기준 그룹화] groupby로 부서 및 이름별 급여 총계 산출")
print(renamed_2)

# 3. 그룹별 여러 집계 연산
grouped_1 = df.groupby('부서') # 분할: 부서 기준으로 그룹 분할
aggregated = grouped_1['급여'].agg(['sum', 'mean', 'max', 'min', 'size', 'count', 'median']) # 적용 및 결합
renamed_3 = aggregated.reset_index().rename(columns={
    'sum': '총_급여',
    'mean': '평균_급여',
    'max': '최고_급여',
    'min': '최저_급여',
    'size': '인원_수',
    'count': '유효_데이터_수',
    'median': '중앙값_급여'
})
print("\n✅ [3. 그룹 기준 여러 집계 연산] groupby로 부서별 급여 집계 연산 산출")
print(renamed_3)
```

## (필수, 추가 챌린지) DataFrame 생성 및 필터링 기능 사용

```python
# DataFrame 생성 및 filtering 기능 활용
# =================== DataFrame 생성 ===================
df = pd.DataFrame(data)
print("✅ 회사 직원 정보 DataFrame 생성")
print(df)

# =================== filtering 기능 활용 ===================
# 1. 그룹화 후 필터링
grouped_1 = df.groupby('부서') # 분할: 부서 기준으로 그룹 분할
filtered = grouped_1.filter(lambda x: x['급여'].mean() > 5000) # 적용 및 결합 # type: ignore
print("\n✅ [1. 그룹화 후 필터링] groupby로 부서별 급여 평균 중 5000 초과 직원 산출")
print(filtered)

# 2. 그룹별 변환: 그룹별 계산 결과를 원래 행에 기존 데이터프레임에 매칭해 반환하는 기능
grouped_1 = df.groupby('부서')
df['급여_평균'] = grouped_1['급여'].transform('mean').round(1)
filtered = df[df['급여_평균'] > 5000]
print("\n✅ [2. 그룹별 변환 후 필터링] groupby로 부서별 급여 평균 컬럼 변환하여 기존 데이터프레임에 매칭")
print(filtered)
```
